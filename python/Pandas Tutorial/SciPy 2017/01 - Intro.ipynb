```python
df = pd.read_csv('https://raw.githubusercontent.com/chendaniely/scipy-2017-tutorial-pandas/master/data/gapminder.tsv', delimiter='\t')
```

<h5> Subsetting Rows </h5>

```python
# loc
# Allows pulling out of rows in DataFrame
# Needs the row label to pull out data
df.loc[0]
```

```python
# iloc
# Index Location
df.iloc[-1]
```

```python
df.loc[[0, 99, 999], ['country', 'lifeExp', 'gdpPercap']]
```

```python
# loc
# Notations to the left of comma subsets rows
# Notations to the right of comman subsets columns
df.loc[df['lifeExp'] > df.lifeExp.mean(), :]
```

<h5> Aggregation of DataFrame </h5>

```python
df.groupby('year')
```

```python
df.groupby('year')['lifeExp'].mean()
```

```python
df.groupby('year')[['lifeExp', 'gdpPercap']].mean()
```

```python
df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean()
```

```python
# .nunique() counts the number of unique values
df.groupby('continent')['country'].nunique()
```
